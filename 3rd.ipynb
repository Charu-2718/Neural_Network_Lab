```python
import matplotlib.pyplot as plt
import numpy as np
def sigmoid(x):
    s=1/(1+np.exp(-x))
    ds=s*(1-s)
    return s,ds
x=np.arange(-6,6,0.01)
sigmoid(x)
fig, ax = plt.subplots(figsize=(9, 5))

ax.spines['left'].set_position('center')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.plot(x,sigmoid(x)[0], color="green", linewidth=3, label="sigmoid")
ax.plot(x,sigmoid(x)[1], color="yellow", linewidth=3, label="derivative")

ax.legend(loc="upper right", frameon=True)
def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    dt=1-t**2
    return t,dt
z=np.arange(-4,4,0.01)
tanh(z)[0].size,tanh(z)[1].size
f, ax1 = plt.subplots(figsize=(9, 5))

ax1.spines['left'].set_position('center')

ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

ax1.plot(z,tanh(z)[0], color="#307EC7", linewidth=3, label="tanh")
ax1.plot(z,tanh(z)[1], color="#9621E2", linewidth=3, label="derivative")

ax1.legend(loc="upper right", frameon=True)
fig.show()
f.show()
```
